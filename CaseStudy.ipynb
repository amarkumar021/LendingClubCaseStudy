Listing down missing data column wise

```python
df.isnull().sum()
```

Listing down columns which is missing data percentage wise in descending order.

```python
missing_value = np.round(df.isnull().sum() * 100/ len(df))
missing_value
missing_value.sort_values(ascending=False)
```

Filtering out set of columns which is missing data more than 30% data.

```python
empty_col = df.isnull().sum()
empty_col = empty_col[empty_col > (0.3 *len(empty_col))]
len(empty_col)
```

Listing down column which misses more than 30% data

```python
empty_col = list(empty_col[empty_col.values > 0.3].index)
empty_col
```

Dropping the columns which misses more than 30% of data.

```python
df.drop(empty_col, axis=1, inplace=True)
```

Validating data frame after cleaning few columns

```python
df.info()
```

### Univariate Analysis

```python
df['term'].value_counts()
```

```python
df['term'] = df['term'].map(lambda x : x.replace('months', ''))
```

```python
df['term'] = df['term'].astype(int)
```

```python
df.info()
```

```python
df['int_rate'].value_counts()
```

```python
df['int_rate'] = df['int_rate'].apply(lambda x : x.replace('%', ''))
```

```python
df['int_rate'] = df['int_rate'].astype(float)
```
